```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as stats
from sklearn.utils import resample
import warnings
warnings.filterwarnings("ignore")
```

```python
Data_of_Attack_Back=pd.read_csv("Data_of_Attack_Back.csv")
Data_of_Attack_Back_BufferOverflow=pd.read_csv("Data_of_Attack_Back_BufferOverflow.csv")
Data_of_Attack_Back_FTPWrite=pd.read_csv("Data_of_Attack_Back_FTPWrite.csv")
Data_of_Attack_Back_GuessPassword=pd.read_csv("Data_of_Attack_Back_GuessPassword.csv")
Data_of_Attack_Back_Neptune=pd.read_csv("Data_of_Attack_Back_Neptune.csv")
Data_of_Attack_Back_NMap=pd.read_csv("Data_of_Attack_Back_NMap.csv")
Data_of_Attack_Back_Normal=pd.read_csv("Data_of_Attack_Back_Normal.csv")
Data_of_Attack_Back_PortSweep=pd.read_csv("Data_of_Attack_Back_PortSweep.csv")
Data_of_Attack_Back_RootKit=pd.read_csv("Data_of_Attack_Back_RootKit.csv")
Data_of_Attack_Back_Satan=pd.read_csv("Data_of_Attack_Back_Satan.csv")
Data_of_Attack_Back_Smurf=pd.read_csv("Data_of_Attack_Back_Smurf.csv")
```

```python
Data_of_Attack_Back_Neptune=resample(Data_of_Attack_Back_Neptune,n_samples=2272,random_state=123)
Data_of_Attack_Back_Normal=resample(Data_of_Attack_Back_Normal,n_samples=5767,random_state=123)
```

```python
Data_of_Attack_Back["Attack"]=1
Data_of_Attack_Back_BufferOverflow["Attack"]=1
Data_of_Attack_Back_FTPWrite["Attack"]=1
Data_of_Attack_Back_GuessPassword["Attack"]=1
Data_of_Attack_Back_Neptune["Attack"]=1
Data_of_Attack_Back_NMap["Attack"]=1
Data_of_Attack_Back_Normal["Attack"]=0
Data_of_Attack_Back_PortSweep["Attack"]=1
Data_of_Attack_Back_RootKit["Attack"]=1
Data_of_Attack_Back_Satan["Attack"]=1
Data_of_Attack_Back_Smurf["Attack"]=1
```

```python
Data_of_Attack_Back_FTPWrite.columns=Data_of_Attack_Back.columns
```

```python
data=pd.concat([Data_of_Attack_Back,Data_of_Attack_Back_BufferOverflow,Data_of_Attack_Back_FTPWrite,Data_of_Attack_Back_GuessPassword,
         Data_of_Attack_Back_Neptune,Data_of_Attack_Back_NMap,Data_of_Attack_Back_Normal,Data_of_Attack_Back_PortSweep
         ,Data_of_Attack_Back_RootKit,Data_of_Attack_Back_Satan,Data_of_Attack_Back_Smurf],axis=0)
```

```python
data.columns=data.columns.str.replace(" ","")
```

```python
def summary(x):
    
    q1=x.quantile(0.25)
    q3=x.quantile(0.75)
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3-1.5*iqr
    
    return pd.Series(data=[x.sum(),x.count(),x.dtypes,x.mean(),x.std(),x.nunique(),lower,upper,iqr,
                           x.quantile(0.01),x.quantile(0.05),x.quantile(0.1),x.quantile(0.25),x.quantile(0.5)
                           ,x.quantile(0.75),x.quantile(0.9),x.quantile(0.95),x.quantile(0.99),x.min(),x.max()],
                    index=["total","count","dtypes","mean","std","nunique","lower","upper","iqr","p1"
                    ,"p5","p10","p25","p50","p75","p90","p95","p99","min","max"])
```

```python
data.apply(summary)
```

```python
cat_col=["protocol_type","service","flag","land","logged_in","root_shell","su_attempted","is_host_login","is_guest_login"]
num_col = data.drop(columns=cat_col + ["Attack"])
```

```python
data_cat=data[cat_col]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in cat_col:
    le=LabelEncoder()
    data[i]=le.fit_transform(data[i])
```

```python
data.head()
```

```python
num_col=num_col.clip(lower=num_col.quantile(0.01),upper=num_col.quantile(0.99),axis=1)
data=pd.concat([data_cat,num_col,data["Attack"]],axis=1)
```

```python
data.apply(summary)
```

```python
X=data.iloc[:,:-1]
Y=data["Attack"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=123)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
model=ss.fit(x_train)
```

```python
xt_train=model.transform(x_train)
xt_train=pd.DataFrame(xt_train,columns=x_train.columns)
xt_train
```

```python
model_test=ss.fit(x_test)
test_xt=model_test.transform(x_test)
test_xt=pd.DataFrame(test_xt,columns=x_test.columns)
test_xt
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params={'n_neighbors':[3,5,7,9],
          'weights':['uniform','distance'],
        'algorithm': ['ball_tree', 'kd_tree']}
```

```python
new_model=GridSearchCV(KNeighborsClassifier(),param_grid=params,cv=5,scoring='roc_auc')
```

```python
new_model.fit(xt_train,y_train)
```

```python
new_model.best_params_
```

```python
train_predic=pd.DataFrame(new_model.predict_proba(xt_train))[1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_train,train_predic)
```
